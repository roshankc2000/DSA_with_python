### Analyzing Algorithms
Suppose we are trying to solve a simple problem of finding the sum of the first 1000 natural numbers. Let's see what our code looks like:

```python
n = 1000
total = 0 
 
 
# iterate n times
for i in range(1, n + 1):
    total = total + i
 
print(total)
```

#### This code takes n steps to complete

Suppose our computer executes one step in one second. In this case, our computer will take 1 * 1000 seconds, or 16 minutes, to run 1000 steps. Nobody wants to wait 16 minutes to run a simple code.

Now, imagine if we want the sum of 1011 natural numbers. It will take 1 * 1011 seconds, which is 1666666666 min 40 seconds.

That is why optimizing algorithms is important so our code is more efficient.

### Optimizing Code
Now, let's try to optimize the code we wrote earlier.

There is a simple formula to calculate the sum of natural numbers.

### Sum = n * (n + 1) / 2
Let's code the sum of n natural numbers again using the formula.

```python
n = 1000
 
total = n * (n + 1) / 2
print(total)
```

#### Assuming it takes 1 second to run 1 step, our optimized code executes in 1 second.

Our previous code correctly found the sum of natural numbers using a for loop, but it took 16 minutes to run.

On the other hand, our optimized code takes only 1 second to run. This is why it is important to calculate the time taken by algorithms.

Even if we want to find the sum of 1011 natural numbers, running our optimized code would take the same amount of time as it would for a sum of 1000 natural numbers.

This makes our code faster and more scalable.

## TIME COMPLEXITY

Introduction
Time complexity is the amount of time an algorithm takes to run (with respect to the size of the input). It is the most important factor to consider while analyzing an algorithm.

But how do we calculate the time complexity of an algorithm?

We could calculate the time complexity of an algorithm based on the execution time. For example, if a code takes 16 minutes to run, its time complexity is 16 minutes.

However, this approach would be wrong because the execution time will be different on different: devices, programming languages, networks, and so on

<br>That's why we calculate the time complexity based on the number of steps required for a given input. For example,

If the time complexity is O(n)

it takes 5 steps for an input of 5
it takes 10 steps for an input of 10
We will cover what O(n) means in detail later in this course.

For now, just remember that the time complexity is calculated based on the number of steps required (regardless of the device or programming language being used).

## Frequency Count Method
We use the frequency count method to calculate the time complexity of an algorithm. The method is based on the number of steps our program takes to execute.

```python
#Here's the same code to find the sum of the first n natural numbers.

def find_sum(n):
    # initialize total to zero
    total = 0
    # iterate n times
    for i in range(1, n + 1):
        # add the number to total
        total = total + i
 
    return total
 
result = find_sum(10)
print(result)
```

## Big O notation

Big Oh (Big O) is a mathematical notation used to represent the time complexity of an algorithm.<br>
Instead of using just plain n for time complexity, we use the O(n) notationread as Big Oh of nto represent it.

Types of Big O
Some common types of Big O notations are as follows:

O(1) - constant time complexity<br>
O(n) - linear time complexity<br>
O(n^2) - quadratic time complexity<br>
O(log n) - logarithmic time complexity

## Constant Time Complexity
An algorithm has a constant time complexity if the execution time does not change with the change in its input size.

Let's see an example.<br>

lst = [22, 23, 24, 25, 11]  STEP = 1<br>
print(lst[3]) # Output: 25  STEP = 2<br>

The above code always takes 2 units of time to execute. Therefore, the time complexity of the above code would be O(2).<br>
However, it is a general practice to use O(1) to represent constant time complexity when it takes a fixed number of steps such as O(2), O(100), etc.<br>

Hence, the time complexity of the above code is O(1).

```python
# example

def get_squares_sum(number1, number2):
    
    # square the numbers
    number1 = number1**2
    number2 = number2**2
    
    # get the sum of squares
    result = number1 + number2
 
    return result
 
# example 
square_sum = get_squares_sum(5, 15)
print(square_sum)
```

In the above program, the get_squares_sum() function takes 4 steps to execute regardless of the input size.

Hence, the time complexity of the above program is O(1).

# linear time complexity
linear time complexity if the execution time is proportional to the size of the input.

This means if the input size increases, the time required to execute the program also increases linearly.

```python
def print_hello(n):
    for i in range(n): # n times
        print("Hello") # n times 
 
print_hello(5) #n times + n times = 2n times hence its linear complexity is O(n)
```

#### Although our code takes 2n steps to run, its time complexity is O(n) and not O(2n). It's because we don't include the coefficient in time complexity, as we have previously mentioned.

# Quadratic Time Complexity

An algorithm has a quadratic time complexity if its execution time is proportional to the square of the input size.

This means if the input size increases, the time required to execute the program increases quadratically.

```python
def print_numbers(n):
    for i in range(0, n):
        for j in range(0, n):
            print(j)
 
print_numbers(2)
```

Although our code takes 
+2^2 steps to run, its time complexity is (^2) using the frequency count method, as we have previously mentioned.

## Note: Quadratic time complexity usually occurs inside nested loops (not always). In such conditions, we can think of different logic to solve the same problem that takes less time to execute.

# logarithmic time complexity

An algorithm has a logarithmic time complexity if the execution time is proportional to the logarithm of the input size.

In an algorithm having logarithmic complexity, if the input size increases, the number of steps required to complete the program grows very slowly.

Before learning about logarithmic time complexity, let's understand logarithms (log) first.

### Note: Generally, in computer science,<br>

#### Due to this slow growth, it is preferable to write algorithms with logarithmic time complexity.

## Example: Logarithmic Time Complexity
#### Let's find out the time complexity of the code below.

```python
def print_numbers(n):
    i = 1
    
    # iterate until i < n
    while i < n:
        print(i)
        # double the value of i each iteration
        i = i * 2
 
print_numbers(5)
```

# Note: Since programs with logarithmic time complexity take fewer steps to execute even for large input, the logarithmic time complexity is preferable whenever possible.

## Problem

Suppose we have a list of integers, num_list, and an integer, target_value. And we need to find out the indices of the two numbers such that they add up to the target_value.

#### Suppose:
num_list = [2, 7, 11, 15]<br>
target_value = 9

The output should be:
[0, 1]

Reason:<br>
2 plus 7 is 9.<br>
The index of 2 and 7 is 0 and 1 respectively.

#### Suppose:
num_list = [2, 7, 11, 15]<br>
target_value = 22

The output should be:
[1, 3]

Reason:<br>
7 plus 15 is 23.<br>
The index of 2 and 7 is 1 and 3 respectively.<br>
Next, we will solve this problem with different complexities.

# Solving the Problem with O(n^2) complexity
The problem can be solved simply by using a nested loop.

```python
def two_sum(num_list, target_value):
    # iterate over each number in the list
    for i in range(len(num_list)):
        # iterate over subsequent numbers in the list
        for j in range(i + 1, len(num_list)):
            # check if their sum equals the target value
            if num_list[i] + num_list[j] == target_value:
                # return the indices of the two numbers
                return [i, j]
 
num_list = [2, 7, 11, 15]
target_value = 9
 
result = two_sum(num_list, target_value)
 
# print the result
print(result)
```

## Reducing Time Complexity of Code

```python
def two_sum(num_list, target_value):
    
    dictionary = {}

    for index, value in enumerate(num_list): 
        # add items of num_list to dictionary
        # value as key and index as value
        dictionary[value] = index
    
    for i in range(len(num_list)):
        complement = target_value - num_list[i]
        
        # check if item's complement is present in dictionary
        if complement in dictionary and dictionary[complement] != i:
            
            # return element index and complement's index
            return [i, dictionary[complement]]
    
num_list = [2, 7, 11, 15]
target_value = 9

result = two_sum(num_list, target_value)
print(result)   
```

Here, the code works similarly to our code on the last page. However, the time complexity of the code is O(n).
